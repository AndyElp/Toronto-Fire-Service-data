##### times are all in the object format so shall be converted to date and time formats

##### incident ward should be int64

##### estimated dollar loss should be changed to float64

```python
# first check time columns to see how they currently appear

tfire [['Ext_agent_app_or_defer_time','Fire_Under_Control_Time', 'Last_TFS_Unit_Clear_Time', 'TFS_Alarm_Time', 'TFS_Arrival_Time']].head()
```

```python
tfire['Last_TFS_Unit_Clear_Time'] =pd.to_datetime(tfire['Last_TFS_Unit_Clear_Time'])
```

```python
tfire ['TFS_Arrival_Time'] = pd.to_datetime(tfire['TFS_Arrival_Time'])
```

```python
# check results

tfire [['Ext_agent_app_or_defer_time','Fire_Under_Control_Time', 'Last_TFS_Unit_Clear_Time', 'TFS_Alarm_Time', 'TFS_Arrival_Time']].head()
```

```python
tfire.dtypes
```

```python
# changing incident ward data type from float64 to int64

tfire['Incident_Ward'] = tfire['Incident_Ward'].astype('int64')
```
